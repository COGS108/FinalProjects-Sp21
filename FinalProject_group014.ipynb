#### Cleaning `Location`
The first two columns we have are both pretty clean. However, `Location` is formatted strangely. Furthermore, we won't need alphabetical characters in the column, as we are only interested in the latitude and longitude values. We will clean this column by removing the text before the latitude and longitude coordinates as well as putting the two coordinates in separate columns. For removing the unnecessary bits of text, we will utilize regular expressions (courtesy of the `re` library).

```python
# Removing everything before parentheses indicating latitude-longitude tuple
crime['location'] = crime['location'].apply(lambda x: re.sub('^[^\(]+', '', x)).str.strip()

# Get list of only numbers in the location column
crime['location'] = crime['location'].apply(lambda x: re.findall(r'\d+\.\d+', x))

# Separate into latitude and longitude columns
crime = pd.concat([
    crime, pd.DataFrame(
        crime['location'].to_list(), 
        columns=['latitude', 'longitude']
    )], 1).drop('location', 1)

# Change to float columns
crime[['latitude', 'longitude']] = crime[['latitude', 'longitude']].astype(float)

crime.head()
```

The `crime` dataset is now in tidy format! Lastly, let's check for missing values.

## Combining Datasets and Adding Features
### `crime` and `scout`
We will not be using conventional DataFrame merging operations on these two datasets, since `scout` is just a column of precincts. Instead, we will append a new column to `crime`, `is_patrolled`, indicating whether a precinct in that dataset is in `scout`. This will be a new categorical feature.

```python
crime['is_patrolled'] = np.where(crime['precinct'].isin(scout['precinct']), 1, 0)
crime['is_patrolled'].hist(ec='black')
plt.title('is_patrolled Precinct Distribution')
plt.xlabel('is_patrolled')
plt.ylabel('Frequency')
```

This column isn't very interesting because only one neighborhood isn't patrolled, and this distribution yields a huge class imbalance. However, as seem in the `scout` precinct distributions, some neighborhoods have more scouting areas than others. We can split these accounts into quartiles, and add a binary column, `patrolled_top_quartile`, to `crime`, to indicate, whether that neighborhood has a number of scouted areas that are in the upper quartile for the whole city.

```python
scout_srs = pd.Series(scout['precinct'])

top_quartile_count = scout_srs.value_counts().describe().loc['75%']
top_quartile_precincts = scout_srs.value_counts()[
    scout_srs.value_counts() >= top_quartile_count
].index.get_level_values(0)

crime['patrolled_top_quartile'] = np.where(crime['precinct'].isin(top_quartile_precincts), 1, 0)
crime['patrolled_top_quartile'].hist(ec='black')
plt.title('is_patrolled_top_quartile Precinct Distribution')
plt.xlabel('is_patrolled')
plt.ylabel('Frequency')
```

The balance isn't as prominent as before, but it's still there, which, taken into real-world context, is good. Note that this column is not an purely an indicator of the DPD's decisions on where to patrol. Since we are measuring number of areas in each precinct, this column can also take into account factors like size.

We have added two new columns based on information from `scouts` into our master dataset.

## Hypothesis Tests on Crime Rates, Trends, and Categories
Given the visualizations and information above, we will conduct hypothesis tests on the following outcome variables:
1. Crime rates (counts)
2. Sharp increases or decreases in crime from 2011 to 2014
3. Distributions of each crime category

These will be based on the following explanatory variables:
1. Patrolled vs. non-patrolled
2. Heavily patrolled vs. non-heavily patrolled
3. Presence of storm infrastructure
4. Master Plan neighborhoods vs. non-Master Plan neighborhoods

### Crime Counts
#### Patrolling Neighborhoods
Here, we use as our outcome variable the **total** amount of crimes from each neighborhood from 2011 to 2014 (i.e., no trends). We will do this by applying another transformation to `neighborhood`.

```python
crime['nb_crime_total'] = crime.groupby('neighborhood')['neighborhood'].transform('size')
crime['nb_crime_total'].head()
```

```python
def stat_by_neighborhood(df, exp_var, outcome):
    """
    Plot a histogram of the outcome variable based on the explanatory variable, 
    return a DataFrame of group statistics
    """
    unique_nb = df.drop_duplicates(subset='neighborhood')
    sns.displot(
        data=unique_nb, 
        x=outcome, 
        hue=exp_var, 
        element='poly', 
        height=6
    )
    plt.title(f'{outcome}: {exp_var}')
    plt.show()
    return unique_nb.groupby(exp_var)[outcome].describe()
    
stat_by_neighborhood(crime, 'is_patrolled', 'nb_crime_total')
```

The two distributions are both heavily skewed right, so most of the crime counts are relatively low. However, the patrolled group has a much larger population, and it's much more varied. The **KS-test is appropriate to use here, even if we have unequal variances and sample sizes**. The hypotheses are specified below.

**Null ($H_0$)**: The alternative cumulative distribution function $F(x)$ of crime counts for non-patrolled neighborhoods *equal* to the empirical cumulative distribution function $G(x)$ of crime counts for patrolled neighborhoods.

**Alternative ($H_1$)**: The cumulative distribution function $F(x)$ of crime counts for non-patrolled neighborhoods is *not equal* to the cumulative distribution function $G(x)$ of crime counts for patrolled neighborhoods.

```python
def ks_test(df, exp_var, outcome, alternative='two-sided'):
    """
    Group by the explanatory variable and return the statistics for a KS
    test on the outcome of each group
    """
    unique_nb = df.drop_duplicates(subset='neighborhood')
    gp_A = unique_nb[unique_nb[exp_var] == 0][outcome]
    gp_B = unique_nb[unique_nb[exp_var] == 1][outcome]
    return ks_2samp(gp_A, gp_B, alternative)


def is_significant(p_val, alpha=0.05):
    """
    Based on p-val from hypothesis test, determine if result was significant
    """
    return 1 if p_val <= alpha else 0

stat, p_val = ks_test(crime, 'is_patrolled', 'nb_crime_total', 'two-sided')
```

We'll maintain a `DataFrame` to keep track of and compare each result of the hypothesis tests. 

```python
results = pd.DataFrame()

def update_results(exp_var, outcome, test, alt, p, reject, idx):
    d = {
        'explanatory_variable': exp_var,
        'outcome_variable': outcome, 
        'test': test,
        'alternative': alt, 
        'p_val': round(p, 3),
        'reject': reject,
        
    }
    results.loc[idx,d.keys()] = d.values()


update_results('is_patrolled', 'nb_crime_total', 'KS test', 'two-sided', p_val, is_significant(p_val), 0)
results
```

We can intuitively visualize the difference between the two samples by creating a **KS goodness-of-fit plot**. It shows the differences between the two cumulative distribution functions 

```python
def get_cdf(grp_A, grp_B):
    two_samples = np.sort(np.concatenate((grp_A, grp_B)))
    cdf_A = [np.round(
        percentileofscore(grp_A, val) / 100, 3) for val in two_samples]
    cdf_B = [np.round(
        percentileofscore(grp_B, val) / 100, 3) for val in two_samples]
    return cdf_A, cdf_B

def plot_two_cdf(exp_var, grp_A, grp_B, title):
    cdf_A, cdf_B = get_cdf(grp_A, grp_B)
    fig = plt.figure(figsize=(10, 7))
    plt.plot(cdf_A, label=f'{exp_var} = 0', c='darkturquoise')
    plt.plot(cdf_B, label=f'{exp_var} = 1', c='lightsalmon')
    for i in np.arange(0, len(cdf_A), 3):
        plt.plot([i, i], [cdf_A[i], cdf_B[i]], c='black', lw=0.4)
    plt.ylabel('CDF')
    plt.legend()
    plt.title(title)
    plt.show()
    
def plot_ks_2samp(df, exp_var, outcome, bin_w=200):
    title = f'{outcome} on {exp_var}: KS Goodness of Fit'
    unique_nb = df.drop_duplicates(subset='neighborhood')
    grp_A = unique_nb[unique_nb[exp_var] == 0][outcome]
    grp_B = unique_nb[unique_nb[exp_var] == 1][outcome]
    plot_two_cdf(exp_var, grp_A, grp_B, title)
    

plot_ks_2samp(crime, 'is_patrolled', 'nb_crime_total')
```

With a $p$-value of 0.451, we will **fail to reject the null hypothesis** that the distributions of total crime from 2011-2014 come from the same population. Now, we will test on the "highly patrolled" explanatory variable.

#### Highly-Patrolled Neighborhoods
**Null ($H_0$)**: The alternative cumulative distribution function $F(x)$ of crime counts for highly (*top quartile*) patrolled neighborhoods *equal* to the empirical cumulative distribution function $G(x)$ of crime counts for non-highly patrolled neighborhoods.

**Alternative ($H_1$)**: The cumulative distribution function $F(x)$ of crime counts for non-highly patrolled neighborhoods is *less* than the cumulative distribution function $G(x)$ of crime counts for highly patrolled neighborhoods.

```python
stat_by_neighborhood(crime, 'patrolled_top_quartile', 'nb_crime_total')
```

```python
def get_results_ks(df, exp_var, outcome, test, alt, idx):
    stat, p_val = ks_test(df=df, exp_var=exp_var, outcome=outcome, alternative=alt)
    results.append(
        update_results(
            exp_var, outcome, test, alt, p_val, is_significant(p_val), idx
        ))

get_results_ks(
    df=crime, 
    exp_var='patrolled_top_quartile', 
    outcome='nb_crime_total', 
    test='KS test', 
    alt='less', 
    idx=1)
results
```

```python
plot_ks_2samp(crime, 'patrolled_top_quartile', 'nb_crime_total')
```

This plot supports our smaller $p$-value, since the `patrolled_top_quartile = 1` is consistently larger than that of `patrolled_top_quartile = 0`. We will **reject the null hypothesis** that the cumulative distribution of total crime for non-highly patrolled neighborhoods is not smaller than that of the highly-patrolled neighborhoods. **This means that values such as the mean and median crime count will be smaller for highly patrolled neighborhoods.**

**What can we conclude from this?** We have sufficient evidence from our KS-tests to say that highly patrolled neighborhoods had less total crime in this dataset from 2011 to 2014, which intuitively makes sense. We cannot the determine the direction of this relationship (i.e, if more police scouting causes less crime, or the other way around), but quantitatively, we can say there is one.
#### Presence of Storm Drains
**Null ($H_0$)**: The cumulative distribution function $F(x)$ of crime counts for neighborhoods with storm drains *equal* to the cumulative distribution function $G(x)$ of crime counts for neighborhoods without them.

**Alternative ($H_1$)**: The cumulative distribution function $F(x)$ of crime counts for highly-patrolled neighborhoods is *not equal to* the cumulative distribution function $G(x)$ of crime counts for non-highly patrolled neighborhoods.

**What can we conclude?** This contravenes our hypothesis that more infrastructure and safer storm protection would yield lesser crime, but we can likely attribute this to the fact that we are only focusing on one sort of infrastructure here, and it's not enough to influence crime rates.

**Alternative ($H_1$)**: The cumulative distribution function $F(x)$ of crime counts for Master Plan neighborhoods is *greater* than the cumulative distribution function $G(x)$ of crime counts for non-Master Plan neighborhoods.

### Crime Trends
We will be adding a column to indicate the percent decrease of crime from 2011 to 2014, with the help of our `nb_crime_rate` column, which indicates the total crimes per year.

```python
def get_pct_change(nb):
    crime_yr = crime[crime['neighborhood'] == nb]['nb_crime_rate']
    return 100 * (crime_yr.iloc[-1] - crime_yr.iloc[0]) / crime_yr.iloc[0]


unique_nb = crime.copy(deep=True).drop_duplicates(subset='neighborhood')
unique_nb['crime_pct_change'] = unique_nb['neighborhood'].transform(get_pct_change)
unique_nb.head()
```

We have sufficient evidence that in our dataset, non-patrolled neighborhoods had a generally larger decreases in crime than did patrolled neighborhoods.

```python
def stat_by_neighborhood_cat(df, exp_var, outcome, category):
    """
    Plot a histogram of the outcome variable based on the explanatory variable, 
    return a DataFrame of group statistics
    """
    sns.displot(
        data=df, 
        x=outcome, 
        hue=exp_var, 
        element='poly', 
        height=6
    )
    plt.title(f'{category}: {exp_var}')
    plt.show()
    return df.groupby(exp_var)[outcome].describe()

nb_crime_unique = crime.copy(deep=True).drop_duplicates(subset=['neighborhood', 'crime_category'])

for c in nb_crime_unique['crime_category'].unique():
    temp = nb_crime_unique[nb_crime_unique['crime_category'] == c]
    print(stat_by_neighborhood_cat(temp, 'is_patrolled', 'category_freq_in_nb', c))
```

Our project aims to find how the factors such as the presence of storm prevention infrastructure, government protection, or policing affect the crime rates among the neighborhoods in the city of Detroit from 2011 to 2014. We use the KS-test to verify the relationship between total crime, crime trend, crime category frequency and the four factors listed above, respectively. Here are the results for the KS-test. For the total crime and patrolled neighborhood, we find that the crime counts for non-patrolled neighborhoods equal to the crime counts for patrolled neighborhoods. Then, testing the crime trend and patrolled neighborhoods infers that non-patrolled neighborhoods have a generally larger decrease in crime than patrolled neighborhoods. It is similar to the relationship between total crime and patrolled neighborhoods that non-patrolled neighborhoods have less crime. We want to emphasize that it is not a casual relationship between crime rates and policing. For the testing of the total crime and neighborhood is highly patrolled, it reveals that highly patrolled neighborhoods had less total crime in this dataset from 2011 to 2014. Also, the KS-test on the crime trend and highly patrolled neighborhoods shows that the percent change in crime for highly patrolled neighborhoods is equal to the one for non-highly patrolled neighborhoods. For the total crime and neighborhood with the storm drains, there isn’t sufficient support from one infrastructure for us to conclude the relationship between the crime count and the presence of storm drains in a neighborhood. The result is consistent with the testing of the crime trend and neighborhood with the storm drains. For the KS-test of total crime and neighborhood with master plan, neighborhoods with master plan have relatively higher crime rate than the neighborhoods without master plan. For the KS-test on the crime trend and neighborhood with master plan, our tests do not confirm either a huge decrease or increase in crime. Finally, we analyze whether any of our factors exacerbate certain categories of crimes and we can compare the earlier results. For the KS-test on the crime category frequency in neighborhood and patrolled neighborhood with master plan, we find that patrolled neighborhoods have significantly higher average rates of aggravated assault and burglary in a neighborhood, which is consistent with the earlier two tests.
