# Overview

Our research project aims to analyze the relationships between poverty, race, and crime to the level of police enforcement in the top 50 most populous cities in the United States. 

Our dataset includes the cities' population, ratio of police employment to city population, ratio of non-white residents to overall population, crime rate, median income, and number of people killed by the police in the city (since Jan 1, 2015). After gathering and cleaning this data, we ended up with a list of all these variables for 43 major US cities due to lack of available data for all top 50 cities. 

Our exploration and analysis suggests that there is a small but statistically significant positive linear correlation between the proportion of police officers to population and the following predictors: ratio of non-white civilians, population, and number of deaths from police shootings. We did not find any statistically signifcant relationships for household median income or crime rate.

# Key Definitions
1. Enforcement ratio: the proportion of police officers to total population in a city
2. Crime rate: the number of crimes per 100,000 civilians in a city
3. Non-white ratio: the proportion of individuals who do not identify as white to total population in a city

# Research Question

What is the relationship between the amount of law enforcement in the top 50 most populous U.S. cities and the following: city population, crime rate, and number of police shootings household, median income and proportion of non-white individuals?

We restricted our analysis to the top 50 most populous cities in the U.S. because general discussion of policing typically centers around major urban areas, which are more heavily populated than other areas (e.g. smaller cities, rural areas). ## Background and Prior Work

Debates over the appropriate sizing of the U.S.’s police force have become increasingly intense in recent years. Activists propose “defunding the police” in order to reallocate funding away from police departments to other government agencies [1]. The intention of reallocating these funds is so that this money can be better spent by non-police organizations to provide safe communities for residents. 

Initially, our group tried to determine which U.S. cities were under- or over-policed. However, we have found that proportionate policing is a vaguely defined term, with the clearest definition proposed by the International City/County Management Association (ICMA). The ICMA uses 3 key criteria for determining proportionate policing [2]:

1. 60 percent of sworn officers in a department are assigned to a patrol function.
2. Officers with a patrol function are utilized at most 60 percent of the time.
3. Each call for service (i.e. 911 call) should receive a police response within 60 minutes.

However, we found it difficult to gather these specific data (for example, the largest repository of 911 call data we found included only 5 cities [3]) and determine what “proportionate policing” entails. Consequently, we revised our research question to analyze the relationship between the ratio of officers to civilians in each city to population, crime rate, median household income, and racial composition, and police shootings. We chose these variables because we believed that they would be predictive of the level of policing in a city.

We did not find any projects that directly analyzed the relationship between police employment and our listed predictor variables, but we encountered projects whose purpose was to provide more insight into the factors that are correlated with the characteristics of a city’s police force.

Police budgets: 
This project from two independent researchers examined the police budgets of 473 U.S. cities. They used a generalized linear model (GLM) to investigate the relationship between a city’s police spending and five predictor variables: population, median income, the Gini index of income inequality, the percentage of people voting Democrat in 2016, and violent crime rate. Their analysis found statistically significant negative correlations between police spending and each of the four predictor variables, with the exception of violent crime rate [4].

Police diversity:
An analysis of the Law Enforcement Management and Administrative Statistics (LEMAS) survey from the Bureau of Justice Statistics by the New York Times provides visualizations of the racial composition of police forces compared to the racial composition of their associated cities. The visualizations point out whether specific cities’ police departments had become “whiter” relative to their city population by 2016, compared to the racial composition from 2013 [5].

# Hypothesis

We believe that there is a positive correlation between a city's police enforcement ratio and its crime rate because a proportionately larger police force is more likely to detect and prosecute crime.

In addition, we believe that lower income areas are more likely to have higher proportions of police to civilians because these areas generally have higher unemployment and may be viewed as “dangerous” or "high-risk" areas by the police. 

Lastly, areas with high concentrations of minority groups are also more likely to be have higher proportions of police to civilians because these areas may be lower-income.

### Police employment: data cleaning and tidying

In addition, note how only the first city in a state has a corresponding state label. The rest have NaN as their value. We will want to rename these to their actual state names.

```python
def fill_states(df):
    # Fill in the NaN values with the correct state
    cur_state = None

    # The first row is guaranteed to start with a string
    for idx, row in df.iterrows():
        if isinstance(row['state'], float):  # float indicates the value is a nan
            df.at[idx, 'state'] = cur_state
        elif isinstance(row['state'], str):  # str indicates the value is a new state
            cur_state = row['state']
        else:
            raise ValueError(f"unexpected value: {(row['state'])}")  # shouldn't be raised
            
fill_states(polemp_df)
```

We want to add a new column for the ratio of the total police enforcement to the population in that city:

```python
polemp_df = polemp_df.assign(enf_ratio = polemp_df['total_emp'] / polemp_df['pop'])
polemp_df.head()
```

Some of the cities are not completely string data, or they may not be the canonical name of that city, because of how the UCR labeled them. We will fix these now:

```python
# Returns a list of cities with numeric digits in them. These were caused by supercripts in the original source data.
def numeric_cities(df):
    cities = list(df['city'])
    numeric = [c for c in cities if any(map(str.isdigit, c))]
    return numeric 

# Returns a list of cities with weird / "non-canon" names. Right now, these are all the cities that have "police" in their
# labels
def non_canon_cities(df):
    cities = list(polemp_df['city'])
    non = [c for c in cities if 'police' in c.lower()]
    return non

print('Cities with digits in them:', numeric_cities(polemp_df))
print('Cities with non-canon labels:', non_canon_cities(polemp_df))
```

```python
def standardize_city(s):
    s = re.sub('[0-9]', '', s)  # Replaces all digits in s with the empty string
    s = s.replace(',', '')  # Replaces all commas with the empty string. 
    # Note: Boston had an extra comma in the offenses data   
    
    # Make city names canonical if they are not
    city_mapper = {
        'Las Vegas Metropolitan Police Department':  'Las Vegas',
        'Metropolitan Nashville Police Department': 'Nashville',
        'Metro Police Authority of Genesee County': 'Genesee County',
        'Camden County Police Department': 'Camden County',
        'Greene County Regional Police Department': 'Greene County',
        'Louisville/Jefferson County metro government': 'Louisville',
        'Nashville-Davidson metropolitan government': 'Nashville'
    }
    
    if s in city_mapper:
        s = city_mapper[s]

    return s 

def standardize_state(s):
    s = re.sub('[0-9]', '', s)  # Replaces all digits in s with the empty string
    s = s.strip()  # Remove leading/trailing whitespace - relevant for median income and city population    
    # Map state to its abbreviation
    if s.title() in STATE_ABBREV:
        s = STATE_ABBREV[s.title()]
    return s 
```

```python
polemp_df['city'] = polemp_df['city'].apply(standardize_city)
assert not numeric_cities(polemp_df)
```

Lastly, let's rename the states to their abbreviations. Source of the State -> abbreviation mapping: https://gist.github.com/rogerallen/1583593

```python
polemp_df['state'] = polemp_df['state'].apply(standardize_state)
```

### Offenses: defining "crime rate"
There isn't a true, universally accepted definition of crime rate. Our analysis will use the FBI's Uniform Crime Reporting program's definition: **the crime rate is the number of crimes per 100,000 civilians.** This definition is commonly used in discussions of the crime rate in U.S. news as well.

Source: https://ucr.fbi.gov/crime-in-the-u.s/2019/crime-in-the-u.s.-2019/topic-pages/expanded-offense

```python
off_df.head()
```

```python
# Sum up the total number of crimes for each city
crime_cols = ['violent', 'kill', 'rape', 'rob', 'assault', 'property', 'burglary', 'larceny', 'motor_theft', 'arson']
off_df = off_df.assign(total_crimes = off_df[crime_cols].sum(axis=1))

# Compute the crime rate - the crime rate uses a factor of 100,000
crime_rate_factor = 100000
off_df = off_df.assign(crime_rate = (off_df['total_crimes'] / off_df['pop']) * crime_rate_factor)
off_df.head()

# Reorder the columns
new_cols = ['state', 'city', 'pop', 'crime_rate', 'total_crimes'] + crime_cols
off_df = off_df[new_cols]
off_df.head(n=3)
```

    - Another limitation of our data is that we could not obtain all the data for each variable for the top 50 populated cities within the United States. For example, the cities Atlanta, Aurora, Fresno, and Philadelphia don’t have crime rate data. This led us to only have data for 43 out of the 50 most populous cities. We also had different data sets for the same variable. For example, the population variable of the top 50 cities was different per dataset. Therefore, we selected what we thought was the most accurate source - in this case, the Census data. 

- Impact:
    - Our project results may be useful reference points for policymakers who are looking to reform city police departments to better serve their populations. The statistically significant relationship between non-white ratio and number of fatal police shootings to enforcement ratio could be points of further investigation (i.e. why do more racially diverse cities tend to have relatively more police officers).

    - In addition, our project contributes to the existing discussion on proportionate policing by providing analysis of variables like non-white-ratio and number of fatal police shootings. Proponents of defunding the police may find that the statistically significant relationship between the number of fatal police shootings and enforcement ratio is an impetus for reform. At the same time, cities are complex, and decisions about police staffing need to be made holistically, on a city-by-city basis, rather than on general principles.
