## Background & Prior Work

Breast cancer will afflict 1 in 8 US women over the course of their lifetime, and is the leading cause of cancer mortality for women. (1, 2) Although breast cancer related mortality has gone down over time due to improved treatment options and early screening, the disease still remains a significant burden on the US healthcare system and on thousands of American women annually. (2) Several factors are associated with increased risk of breast cancer, including older age, genetic predisposition, age at menopause, exposure to hormones (both internal and external sources), and lifestyle factors (2). Lifestyle factors that can promote or prevent breast cancer are of most interest to physicians, as they can more easily be controlled through intervention by the patient. These factors may include obesity, alcohol consumption, exercise frequency, and diet. (2) 

In general, there are three main types of breast cancer. Hormone receptor positive breast cancers have an upregulation of either estrogen receptors, progesterone receptors, or both. (3) These cancers have the best prognosis and are typically treated with hormone therapy that targets these upregulated receptors. HER-2 positive breast cancers have an upregulation of the HER-2 growth receptor, and can be treated using anti-HER2 antibody treatments such as Herceptin. (3) Lastly, triple-negative breast cancer is neither ER+, PR+, or HER2+, and is most common amongst women with a BRCA1 gene mutation. (3) Due to the lack of targetable receptors, this type of breast cancer tends to have the worst prognosis. 

Pinpointing the exact mechanisms that directly lead to cancer is not yet a possibility, but the influence of diet has been an increasingly common consideration in the study of breast cancer development and prognosis. (4) Consumption of certain types of foods may be linked to various chemical interactions within the body that can either increase or decrease risks of breast cancer. (4) Correlative studies are working to establish a cause-and-effect relationship between nutrition and breast cancer, but definitive conclusions have yet to be drawn. (4)

Soy in particular has often been the subject of these correlative studies, especially for hormone-receptor positive breast cancers. The hormone estrogen has been linked to the development of breast cancer, and soy products contain the compound genistein, which has weak estrogenic effects. (5,6,7) Accordingly, one should expect that there should be a correlation between the occurrence and development of breast cancer and a high consumption of soy products. However, studies thus far have been unable to confidently link soy dietary patterns directly to estrogen levels and increased breast cancer risk. (5) In fact, there are other studies which suggest the opposite: soy may actually induce a protective effect against recurring breast cancer, rather than exacerbating it. (6,7) Spurred by these unanswered questions, we wish to investigate both dietary patterns and corresponding breast cancer prognosis within patients of a clinical trial.

To better understand the relationship between diet and cancer prognosis, we will analyze data from a clinical trial conducted at UC San Diego between 1995 and 2006. The Women’s Healthy Eating and Living (WHEL) study followed 3088 women with breast cancer and randomized them to either intensive diet intervention or comparison groups to determine whether diet had a significant impact on overall survival and prognosis. (8) Although this dataset will not enable us to study the effects of soy consumption on the risk of developing breast cancer, it will allow us to determine whether there is a link between soy and cancer progression and survival rates post-diagnosis.

# Hypothesis


We hypothesize that groups of individuals with a high soy diet experience a different prognosis after breast cancer diagnosis compared to groups who consume less soy. Due to supposed interactions between genistein and estrogen receptors associated with breast cancer growth, it can be inferred that soy consumption increases risk of breast cancer development and worsens prognosis. However, due to conflicting studies that argue for the protective effects of soy products, we will mainly be focused on whether soy induces a change in either direction (having either a positive or negative effect).

# Data Analysis & Results

## The Relationship Between Diet and Breast Cancer

To observe as to whether consumption of soy or alcohol has lead to poorer patient outcome, the endpoint dataframe was merged with the Year 0 nutrition dataframe, and additional data cleaning was performed.

*In the following data:*

For Vitality:

- 0 = Dead
- 1 = Alive
- 2 = Unknown

For BC Status:

- 0 = No Evidence of Recurrence
- 1 = Confirmed – New Primary Breast Cancer
- 2 = Confirmed – Local Recurrence
- 3 = Confirmed – Regional Recurrence
- 4 = Confirmed – Distant Recurrence

For BC Death:

- -1 = Not Dead
- 0 = Dead from a cause other than Breast Cancer
- 1 = Dead from Breast Cancer
- 2 = Dead from Cancer, not confirmed breast but likely so

```python
## Further cleaning of initial nutrition data
# drop all but Patient ID, alcohol consumption, and soy
df_ndsy0 = df_ndsy0.drop(columns=['Kcal', 'Protein (g)', 'Total Fat (g)', 'Total Carb (g)', 'Non-soy Legume'])
df_ndsy0.columns = ['id', 'alc', 'soy']

# convert values in these columns to floats
df_ndsy0['alc'] = df_ndsy0['alc'].apply(float)
df_ndsy0['soy'] = df_ndsy0['soy'].apply(float)

# Create boolean columns for soy and alcohol
df_ndsy0['alcbool'] = df_ndsy0['alc'] > 0
df_ndsy0['soybool'] = df_ndsy0['soy'] > 0
df_ndsy0.head()
```

```python
# Further cleaning of endpoint dataframe
df_end2 = df_end.drop(columns=['Grade', 'Stage', 'Recurrence'])

# Rename columns for shorthand names
df_end2.columns = ['id', 'vit', 'bcstatus', 'bcdeath', 'yd2r', 'yd2d']

# Merge Y0 consumption dataframe with endpoint dataframe
df_cons_end = df_ndsy0.merge(df_end2, how = 'inner', on = 'id')
df_cons_end.head()
```

Because of the small number of soy consumers and alcohol consumers in this data set, we had two options on how to analyze this data: either compare consumers with non-soy consumers or only analyze the patient outcome of consumers only. We will first attempt the former analysis method here.

### Comparing Patient Outcome Between Soy and Non-soy Patients

```python
## Plot proportion of vitality, recurrence, and death for soy patients and non-soy patients

# Create proportion dataframes
prop_df1 = df_cons_end.groupby(['soybool'])['vit'].value_counts(normalize=True).rename('proportion').reset_index()
prop_df2 = df_cons_end.groupby(['soybool'])['bcstatus'].value_counts(normalize=True).rename('proportion').reset_index()
prop_df3 = df_cons_end.groupby(['soybool'])['bcdeath'].value_counts(normalize=True).rename('proportion').reset_index()

# Plot proportional barplots
fig, axes = plt.subplots(3, figsize = (10,18))

sns.barplot(x='vit', y='proportion', hue='soybool', data=prop_df1, ax=axes[0])
axes[0].set(xticklabels = ['Dead', 'Alive'])
axes[0].set_title('Vitality and Soy Consumption')

sns.barplot(x='bcstatus', y='proportion', hue='soybool', data=prop_df2, ax=axes[1])
axes[1].set_title('Recurrence and Soy Consumption')
axes[1].set(xticklabels = ['No recurrence', 'New Cancer', 'Local', 'Regional', 'Distant'])

sns.barplot(x='bcdeath', y='proportion', hue='soybool', data=prop_df3, ax=axes[2])
axes[2].set(xticklabels = ['Alive', 'Death (not BC)', 'Death (BC)', 'Death (unknown cause)'])
axes[2].set_title('Death Cause and Soy Consumption')

fig.suptitle('Soy Consumption and Patient outcome', fontsize=16);
```

Visually, we find that across these three qualitative metrics of patient outcome, whether soy is eaten or not causes little effect on the progression of breast cancer. To see if there are any differences among the soy patients, we will isolate that group and analyze their data as a separate population.

### Isolating Soy Patients and Analyzing Correlation between Soy and Patient Outcome

After extracting a subset of the data containing only soy consumers, the patients were further filtered by whether they experienced a recurrence in their cancer over the course of the survey period. Then, a scatterplot of Years to Recurrence vs. Soy Consumption amount was created.

```python
# Extract Soy-Only Dataset
df_soy = df_cons_end[df_cons_end['soybool'] == True]

# Extract patients that experienced Recurrence
df_soy_rec = df_soy[df_soy['bcstatus'] != 0]

# Plot scatterplot btwn soy consumption and year to recurrence
g1 = sns.scatterplot(x='soy', y='yd2r', data=df_soy_rec)
g1.set_title('Year to Recurrence vs. Soy Consumption');
```

Visually, there doesn't seem to be a clear correlation between these two variables. To confirm this, linear regression was performed, with soy consumption as a predictor for years to regression.

```python
# Perform linear regression btwn soy consumption and year to recurrence
outcome1, predictors1 = patsy.dmatrices('yd2r ~ soy', df_soy_rec)
model1 = sm.OLS(outcome1, predictors1)
results1 = model1.fit()
print(results1.summary())
```

With a very low R-squared and a high p-value, we fail to establish a correlation between soy and recurrence

We will repeat this process again with patient death.

```python
# Extract patients that died
df_soy_death = df_soy[df_soy['vit'] == 0]

# Plot scatterplot btwn soy consumption and year to death
g2 = sns.scatterplot(x='soy', y='yd2d', data=df_soy_death)
g2.set_title('Year to Death vs. Soy Consumption');
```

Again, there is no clear correlation between soy and years to death.

```python
# Perform linear regression btwn soy consumption and year to death
outcome2, predictors2 = patsy.dmatrices('yd2d ~ soy', df_soy_death)
model2 = sm.OLS(outcome2, predictors2)
results2 = model2.fit()
print(results2.summary())
```

Linear regression results show a low R-squared value and a large p-value. Thus, we fail to reject the null regarding the effects of soy consumption on patient outcome.

### 1. Possible Lifestyle Confounding Variable: Weight

After identifying the relationship between diet and breast cancer prognosis in a population of women in Southern California, possible confounding variables must be explored. Identification of confounding variables can help clarify whether the results we found were affected by alternate variables that would otherwise be buried in our analysis.

The following EDA examines the distribution of the Lifestyle Dataset observations in regards to weight. Because we are looking for a relationship between consumption of particular substances and breast cancer progression, it is possible that weight may act as a confounding variable. Therefore, it is worthy to explore and visualize the information given on weight. 

```python
df_ndsy0.columns = ['ID', 'alc', 'soy', 'alcbool', 'soybool']
df_life2 = df_life.merge(df_ndsy0, how = 'inner', on = 'ID')
sns.histplot(data = df_life2, x = 'Weight', kde = True)
plt.title('Distribution of Patient Weight');
plt.xlabel('Weight (lbs)');
```

The provided histogram shows a distribution of weight skewed right. This is not an unusual outcome, as many patients of varying age, height, and background will have weights between 100 and 200 pounds, as shown. Often, some patients will weigh more than 200 pounds. However, significantly higher weights may indicate obesity and other health complications that are weight-induced. We will thus include the height and BMI variables from the Medical dataset in this Lifestyle EDA.

```python
# Merge Height and BMI from df_med to df_life in df_lifeEDA
df_lifeEDA = df_life2.merge(df_med, how = 'inner', on = 'ID')
df_lifeEDA = df_lifeEDA.drop(columns = ['Pre-BC Weight', 'Age18 Weight', 'Max Weight', 'Time at Max Weight',
                                   'Insurance Change?', 'Lump/Mast', 'Radiation','Chemo','Tumor Type',
                                   'Estr Recep','Prog Recep','Stage','Tumor Size','Weight_y', 
                                        'alc', 'soy', 'alcbool', 'soybool'])

df_lifeEDA2 = df_lifeEDA.merge(df_end, how = 'inner', on = 'ID')
df_lifeEDA2 = df_lifeEDA2.drop(columns = ['Vitality', 'Grade','Stage', 'Recurrence',
                                          'Years Diagnosis to Recurrence', 'Years Diagnosis to Death'])
# Dropped variables are not necessarily being removed from analysis. Further EDA will be performed pertaining to df_med.
# Vitality is being dropped because BC Death is inclusive and more descriptive of that information.
# Recurrence is being dropped because BC Status is inclusive and more descriptive of that information.

fig, axes = plt.subplots(1,2, figsize = (9,5))
sns.histplot(data = df_lifeEDA2, x = 'BMI', hue = 'BC Status', kde = True, palette='coolwarm', ax = axes[0])
sns.histplot(data = df_lifeEDA2, x = 'BMI', hue = 'BC Death', kde = True, palette='coolwarm', ax = axes[1])
fig.suptitle('BMI Breakdown of Patients Grouped by BC Status/Death');
```

The graphs above are provided to show visually, it appears that BMI, BC Status, and BC Death follow the same trends. This suggests no obvious relationship between BMI and the breast cancer wellness of a patient across the full BMI range. However, having a significantly high BMI could induce additional health complications that may affect the breast cancer status of a patient. While there may not be any obvious relationship across the entire range of BMI, there may be a significant difference between BC outcome of obese and non-obese patients. To test this, a p-value test is most appropriate.

```python
# BC-caused DEATH comparison between obese and non-obese patients
df_lifeEDA2_obese = df_lifeEDA2[df_lifeEDA2['BMI']>=30] #According to CDC, BMI of 30.0 and above is obese
df_lifeEDA2_nonobese = df_lifeEDA2[df_lifeEDA2['BMI']<30]
t_val_obesedeath, p_val_obesedeath = ttest_ind(df_lifeEDA2_nonobese['BC Death'], df_lifeEDA2_obese['BC Death'])
if p_val_obesedeath < 0.01:
    print('There is a significant difference between BC-caused deaths of obese and non-obese patients!')
else:
    print('There is NOT a significant difference between BC-caused deaths of obese and non-obese patients!')
```

```python
# BC RECURRENCE comparison between obese and non-obese patients
t_val_obeserec, p_val_obeserec = ttest_ind(df_lifeEDA2_nonobese['BC Status'], df_lifeEDA2_obese['BC Status'])
if p_val_obeserec < 0.01:
    print('There is a significant difference between BC recurrence of obese and non-obese patients!')
else:
    print('There is NOT a significant difference between BC recurrence of obese and non-obese patients!')
```

Running p-value analysis shows a significant difference between BC-caused deaths in obese patients and BC-caused deaths in non-obese patients. On the other hand, the same analysis shows an insignificant difference between BC recurrence in obese patients and in non-obese patients.

**Therefore, our analysis shows that whether a patient is obese (BMI => 30) or not acts as a confounding variable between diet and BC-caused deaths. Neither obesity, BMI, nor weight act as a confounding variable between diet and BC recurrence.**

In this project, we aimed to investigate whether there was a relationship between a patient’s diet and their cancer prognosis, with the aim of potentially improving patient response to treatment and overall survival. Breast cancer is one of the most common forms of cancer (affecting 1 out of every 8 US women in their lifetimes), so we focused our efforts on breast cancer tumors. Current scientific research suggests that soy (which mimics estrogen in the body) and alcohol (a known carcinogen, which can cause DNA damage), both increase cancer risk and cancer progression. As such, using data from a study conducted at UC San Diego, we aimed to determine whether there was a link between consumption of soy or alcohol and increased metastatic activity for breast cancer patients. 

This data analysis failed to find a significant link between consumption of soy or alcohol and worsened cancer prognosis in a study of Southern California women with breast cancer, as demonstrated through scatter plot regressions with R^2 value of zero. Attempts to analyze this dataset were hindered by lack of responses to all diet questionnaires by patients, and the fact that only a small minority of the patients were consuming any amount of either soy or alcohol while enrolled in the study (approximately ⅓ for each). Additionally, we identified key confounding variables that impacted the prognosis of patients, including whether or not they recurred (had cancer come back a second time), whether they were obese (p-value of 0.002), and their ethnicity (with African-American, Pacific Islander, and Mixed/Other patients having significantly higher rates of breast cancer related mortality than other groups). Although these differences may be due to differing consumption patterns across these groups, it is much more likely that these patterns are due to disparities in care and systemic racism that affects patient health. 

Although we failed to identify a significant diet component that affects cancer prognosis, this analysis still has the potential to impact patient care and future medical research. We highly suggest more clinical studies to better understand the relationship between diet and cancer, expanding beyond soy and alcohol to explore relationships between vitamin consumption, protein to fats ratio, and other metabolics information. This information may benefit both patients and physicians in the future, to best guide patients towards a diet that can improve their prognosis. In addition, we have highlighted obesity and ethnicity as two variables that do have an impact on breast cancer prognosis, and highly encourage future analysis of these patient populations to determine the root causes of these disparities as well as the best ways to combat them. Future analysis of a larger, more diverse patient population will allow for a more accurate analysis of the relationship between diet and breast cancer prognosis, and potentially improve the lives of patients suffering from this disease.
