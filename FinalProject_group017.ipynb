Our group is interested in finding whether hip hop has declined in lyrical complexity since the 1990s. We examined the lyrics of approximately 100 of the most popular hip hop songs since 1990 and measured the lyrical complexity using repetition, average word length, and a rhyme score designed to quantify the quality of rhyme. ## Background & Prior Work

In a 2015 article titled “The Lost Art of Lyricism”, Wu-Tang Clan founding member GZA claimed that modern hip hop has become repetitive, and that the complexity and ingenuity of rhymes have declined since the 80’s and 90’s (1). In the years since, artists such as Eminimem (2), J Cole (3), and Ice Cube (4) have echoed the same claim of declining lyrical complexity in hip hop. This sentiment has also become generally popular in the hip hop community, with the late 80’s and 90’s being referred to as the “Golden Age of Hip Hop”, when the genre was considered to be at its most innovative (5). However, newer artists such as Lil Yachty have responded by arguing the complex lyrics are not necessary for hip hop to be enjoyable, and that more lyrical music is still being created by popular artists such as J. Cole and Kendrick Lamar (6).

A 2019 study of lyrical diversity in Digital Humanities Quarterly found that lyrical diversity (measured by average number of words and ratio of total words to unique words) of popular music as a whole increased sharply in 1991, corresponding with the rise of hip hop. This study also found that these measures of lyrical diversity peaked in the early 2000’s but began to rise again after the mid 2010’s (7). A similar analysis of pop music using a metric similar to ratio of total words to unique words also found an increase in the early 90’s followed by a decline and then slight increase post 2015 (8). The author hypothesized that the decline may be due to hip hop’s slow transformation into pop music, since it was the main driver for increased lyrical complexity in the first place.

An analysis in Northeastern University’s StoryBench also measured lyrical complexity over time using the number of unique words in each song, but only with hip hop music. The analysis found that the 90's had the highest mean while the 2000’s had the lowest, although the difference across all three were fairly slight (9). However, all three analyses failed to account for how the complexity of rhyme has changed over time. Rhyme is crucial to Hip Hop lyrics, a defining element of the genre. The declining ingenuity of rhyme was also a large part of GZA’s argument as to why lyricisim has declined. Therefore, our analysis will not only focus on the complexity of the words themselves, but also the quality of rhyme present in the lyrics. Thus, we will use three metrics for lyrical complexity: repetition (number of total words divided by number of unique words), average word length, and a rhyme score that quanitifies the quality of rhyme (further detailed in our analysis).

We take the `release_date` column of `tracks_df` and format each entry such that the month and day are removed, leaving only the year of release.

That is because in the artist list, the first entry is the main artist and all the subsequent entries are the artist(s) featured on the song. And we want to avoid searching for songs on Genius with featured artists instead of the main artist.

```python
def pick_first(artists):
  return literal_eval(artists)[0]

tracks_df['artists'] = tracks_df['artists'].apply(pick_first)
tracks_df
```

We take `tracks_df` and remove any extra variables that are not essential for the sake of the research question. We also rename the columns in order to simplify merging this dataset with others.

```python
# converts all letters in strings to lowercase
tracks_df['name'] = tracks_df['name'].str.lower()

# replaces accented letters with non-accented equivalents
tracks_df['name'] = tracks_df['name'].str.normalize('NFKD')\
       .str.encode('ascii', errors='ignore')\
       .str.decode('utf-8')
```

### Merging `artists_country_df` and `tracks_df`

We now merge `artists_country_df` (from the Spotify Artists Dataset and the Country of Artists Dataset) and `tracks_df` (from the Spotify Tracks Dataset) on `name` column from both datasets in order to assign each artist to their corresponding tracks and place the data in `track_merge_df`. We merge these datasets using an inner join.

```python
track_merge_df = artists_country_df.merge(tracks_df, on = 'name', how = 'inner')
print("Number of Tracks: ", track_merge_df.shape[0])
track_merge_df.head()
```

We drop all duplicate song titles in order to guarantee that we do not mismatch song lyrics when pulling from Genius. We also change every entry in `release_date` to be a number instead of string so that we can then only include songs that were created on or after 1989.

```python
track_merge_df = track_merge_df.drop_duplicates(subset=['song_name'], keep = 'first')
track_merge_df['release_date'] = pd.to_numeric(track_merge_df['release_date'])
track_merge_df = track_merge_df[track_merge_df['release_date'] >= 1989]
print("Number of Tracks: ", track_merge_df.shape[0])
track_merge_df.head()
```

We normalize all names in `track_merge_df` to a common naming convention that will be utilized across all datasets to avoid future discrepancies while merging datasets.

```python
# converts all letters in strings to lowercase
track_merge_df['song_name'] = track_merge_df['song_name'].str.lower()

# replaces accented letters with non-accented equivalents
track_merge_df['song_name'] = track_merge_df['song_name'].str.normalize('NFKD')\
       .str.encode('ascii', errors='ignore')\
       .str.decode('utf-8')
```

We then take each song title that mentions a featured artist and remove that specific part of the title while retaining the main title. We then remove specific odd instances of issues with various song titles seen throughout the dataframe. We also remove any entry where the song title is not available.

```python
def removeFeat(song_name):
  if 'feat' in song_name or 'ft.' in song_name or '(with' in song_name:
    song_name = re.sub(r'\([^)]*\)', '', song_name)
    song_name = re.sub(r'\[.*?\]', '', song_name)
    song_name = song_name.strip()
  return song_name

def removeExtraData(song_name):
  if 'B-Boy Bouillabaisse: ' in song_name:
    song_name = song_name.replace('b-boy bouillabaisse: ', '')
  if '()' in song_name:
    song_name = song_name.replace('()', '')
  if song_name == '-':
    song_name = song_name.replace('-', '')
  if song_name == ',':
    song_name = song_name.replace(',', '')
  if song_name == '!!':
    song_name = song_name.replace('!!', '')
  if song_name == '!':
    song_name = song_name.replace('!', '')
  return song_name

track_merge_df.dropna(subset=['song_name'], inplace=True)
track_merge_df['song_name'] = track_merge_df['song_name'].apply(removeFeat)
track_merge_df['song_name'] = track_merge_df['song_name'].apply(removeExtraData)
print("Number of Tracks: ", track_merge_df.shape[0])
```

We remove all songs that are likely duplicates of already existing songs in the dataframe. For example, we remove any song that is a remastered version of an already existing song. We do this for various types of songs in order to minimize the number of songs that we will have to eventually feed to the Genius API.

```python
track_merge_df['song_name'] = track_merge_df['song_name'].str.strip()
track_merge_df = track_merge_df[~(track_merge_df['song_name'] == '')]
track_merge_df['name'] = track_merge_df['name'].str.strip()
track_merge_df = track_merge_df[~(track_merge_df['name'] == '')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('remaster')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('version')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('live')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('edited')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('soundtrack')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('skit')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('remix')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('edit')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('radio')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('includes')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('mix')]
track_merge_df = track_merge_df[~track_merge_df['song_name'].str.contains('instrumental')]
print("Number of Tracks: ", track_merge_df.shape[0])
track_merge_df.head()
```

##Measuring Repetitiveness of Hip-Hop Songs over Time

We measure the repetitiveness of a song by counting the total number of words in the song and dividing it by number of unique words in the song. This means that the more repetitive the artist is with their lyrics, the higher the song's repetitiveness metric will be. A higher repetitiveness metric means lower lyrical complexity of the song. That is because the more repetitive a song is, the less unique words, unique rhymes and unique wordplay it will contain. Therefore, its complexity will be lower.

We use the function `calc_rep_metric` to calculate the repetitiveness metric of a song.

```python
def calc_rep_metric(lyrics_tokens):
  if len(lyrics_tokens) == 0:
    return np.nan
  unique_words = Counter(lyrics_tokens).keys()
  rep_metric = len(lyrics_tokens)/len(unique_words)
  return rep_metric
```

We apply the `calc_rep_metric` function on tokenized lyrics **with** stopwords. This is to account for the fact that artists often use stopwords instead of meaningful syllables to fill up the syllable space in their verses. We want to account for these filler stopwords in our repetitive metric. As the more stopwords that are being used to fill a hip-hop song, the more repetitive, and less lyrically complex it will get. 


We store the repetitiveness metric of the songs in the `rep_metric` column.

```python
all_df['rep_metric'] = all_df['lyrics_tokens'].apply(calc_rep_metric)
```

## Measuring Average Word Length in Hip-Hop Songs over Time

The average word length of a song is simply the average of the lengths of all words used in that song's lyrics. The higher the average word length of a song, the more lyrically complex it is. That is because, the longer the word is, the more syllables it will generally contain. And, it requires more ingeniuty to rhyme larger number of syllables and make them flow. As an example, the rhyming word pair of ['mockery', 'lottery'] is more complex relative to the rhyming word pair of ['cat', 'hat'] because the first pair rhymes 3 syllables ([mo-, lo-], [-cker, -ter], [-ry, -ry]) compared to only 1 rhyming syllable in the second pair. Therefore, lyrics containing longer words are generally more complex.

We use the function `calc_average_word_length` to calculate the average word length of a song.

```python
def calc_average_word_length(sentence):
  total_char_count = 0
  for word in sentence:
    total_char_count += len(word)
  return total_char_count / len(sentence)
```

When calculating the average word length of a song, we do not filter out the stopwords in the song lyrics. This is because in hip-hop songs, a longer word with multiple syllables makes the lyrics more complex regardless of whether it is a stopword or not. Consider the following two verses:

**Verse B**<br>
Copped another Gucci just like one I had,<br>
Showed my Gucci to my opp and now he mad,<br>
Turned around and I saw his baby-boo,<br>
When she saw my Gucci she said I love you,<br>

From the above two verses, Verse A will be considered lyrically more complex because it has two rhyming syllables in the end [should-n't, wood-en, could-n't, wool-en], compared to only one rhyming syllable in Verse B's ends [had, mad, -boo, you]. And it does not matter if the rhyming words are stopwords i.e. [shouldn't, couldn't, had, you]. They add to the lyrical complexity of the verse.

Thus, we apply the `calc_average_word_length` function on tokenized lyrics **with** stopwords.

## Measuring Rhyme Scores of Hip-Hop Songs over Time

We measure the Rhyme Score of a song by counting the total number of words in the song that rhyme together. Since Hip-hop lyrics are mainly about rhyming words together, the higher the rhyme score, the more lyrically complex the lyrics are. 


We also take into account how many of these rhyming words rhymed with each other. For instance, consider the following two verses:

**Verse A**:

*I'm just tryna rap, <br>
All we do is trap, <br>
I'm not gonna cap, <br>
We gon' get her snap,*

**Verse B**:

*I'm just tryna rap, <br>
All we do is trap, <br>
She thinks that I'm hot, <br>
We be smoking pot,*

Note that verse A contains the rhyming words [Rap, Trap, Cap, Snap], while verse B contains the rhyming words [Rap, Trap, Hot, Pot]. Although both verses have the same number of rhymes, verse A has rhyming words that rhyme with each other more, (all four rhyming words rhyme with each other). Therefore, verse A will get a higher rhyme score relative to verse B. We make this choice because it requires more lyrical ingeniunity to come up with more rhymes using the same syllable, and it also improves the flow of the song as their is less change in the rhyme sounds. Therefore, a scheme in which rhyming words rhyme with each other more are considered more complex lyrically and we will give it a higher score.   

Also, our rhyme score merely counts the number of rhyming words and does not account for the proportion of rhyming words in the lyrics i.e. if song A is 200 words long and has 10 rhyming words while song B is 20 words long and has only 1 rhyming word (i.e. essentially no rhymes), song A will get a higher rhyme score because it has more rhyming words even though the the proportion of rhyming words in both songs were the same. We make this choice because it requires more lyrical skill to make longer songs rhyme as they contain more words, therefore credit must be given to a song which mantains a similarly complex rhyme scheme but for more words. 

We use the function `calc_rhyme_score` to calculate the rhyme score of a song.

```python
def calc_rhyme_score(words_list):
  if len(words_list) == 0:
    return np.nan
  uniq_words = set(words_list)
  diff_rhymes = 0
  total_rhyme_count = 0
  while len(uniq_words) > 0:
    curr_len = len(uniq_words)
    curr_word = uniq_words.pop()
    curr_rhymes = pronouncing.rhymes(curr_word)
    uniq_words = uniq_words.difference(curr_rhymes)
    rhyme_count = curr_len - len(uniq_words)
    total_rhyme_count = total_rhyme_count + rhyme_count
    diff_rhymes = diff_rhymes + 1
  return total_rhyme_count / diff_rhymes
```

When calculating the rhyme scores of a song, we do not filter out the stopwords in the song lyrics. This is because of the same reason that we did not filter stopwords when calculating the average word length of the songs i.e. rhyming syllables makes the lyrics more complex regardless of whether it is a stopword or not. 

Thus, we apply the `calc_rhyme_score` function on tokenized lyrics with stopwords.

```python
all_df['rhyme_metric'] = all_df['lyrics_tokens'].apply(calc_rhyme_score)
```

Ensuring that our outliers are adequately explained and that they do not connote any discrepencies in our data, we move forward with our Linear Regression Model.

```python
outcome, predictors = patsy.dmatrices('rhyme_metric ~ year_norm', df_years)
mod = sm.OLS(outcome, predictors)
res = mod.fit() 
print(res.summary())
```

From the Linear Regression Results, we see that there is a negative correlation between release years and rhyme score of hip-hop songs just like we hypothesized. The magnitude of the relationship is 0.0015 i.e. every year, the rhyme score of hip-hop songs decrease by 0.0015. The p-value for our model is 0, which suggests that our findings are significant and that they did not arise by mere chance. But the R-squared value is 0.418 which connotes that only 42% of the decrease in rhyme score of hip-hop songs are explained by the release year of the song. This suggests that the correlation between the rhyme score of a hip-hop song and its release date is relatively weak.

# Conclusion & Discussion

Our graphs reveal a slightly downward trend in lyrical complexity since 1990, which _visually_ supports our hypothesis. However, our numerical results show that time, as an independent variable, only explains less than `50%` of hip-hop's lyrical complexity decay from 1990 onwards for each of the three metrics measured. This means that there are other variables such as song length that could further explain the decrease in hip-hop's lyrical complexity. 

However, our project was not without its limitations. For instance, the `pronouncing` package that we used to find rhymes does not contain slang, which is very common in Hip Hop lyrics.  In addition, when calculating rhyme score, we count all instances of rhyming words in the lyrics. However, the presence of rhyming word pairs does not necessarily mean the artist actually rhymed them in the lyrics. As seen in the following verse, cat and hat would be incorrectly counted as a rhyme:  I got a hat on my head, there is a skinny black cat. Furthermore, although we used average word length as a measure of lyrical complexity, it is not always the case that longer words are more complex. For example, Hip Hop artists tend to use long ad-libs (e.g skrrrrrt) that do not add to the lyrical complexity. We also saw that the Genius API is not fool-proof, sometimes providing text that are not even lyrics when the song is not found. Although we removed some of these cases when they appeared as outliers, we could not guarantee that all instances were removed. Despite only including American and Canadian artists, we still had non-English songs which we did not account for in our analysis. Finally, we can not apply our results to hip hop globally since we removed non American and Canadian artists.

Our results show that there is indeed some truth to legendary hip hop artists Eminem, Dr. DRE, and Ice Cube's claims of declining lyrical complexity in hip-hop. However, we also saw that time plays a fairly minor role in this decline. There are other factors that potentially contribute more to this decline. Therefore, if we are able to pinpoint these other factors, that will give us a better chance of reviving, or potentially even exceeding the lyrical ingenuity presented in the 1990s, an era which, to this day, is considered by many to be the “Golden Age of Hip-Hop”. It is true that lyrical complexity is not necessary for hip-hop to be enjoyable. However, the complexity of lyrics, the ingenuity of rhymes, is what separates hip-hop from other music genres. It not only makes hip-hop unique, it makes it art. Thus, strong lyrical complexity is important for hip-hop’s societal perception as a legitimate art form. Encouraging its growth can help bring new fans, increase innovation in the genre, and help bring about a new Golden Age of Hip-Hop.
