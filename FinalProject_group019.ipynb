### Initial Cleaning and Merging Datasets
Load [poster](https://www.kaggle.com/neha1703/movie-genre-from-its-poster) and [summary](https://www.kaggle.com/jrobischon/wikipedia-movie-plots) dataset.

```python
poster_path = "data/raw/MovieGenre.csv"
summ_path = "data/raw/wiki_movie_plots_deduped.csv"
```

```python
def extract_date(x):
    """
    Extracts date from title column of poster dataset.
    """
    d = re.findall('\(\d{4}\)', x)
    if len(d) > 0:
        return int(re.sub('[()]', '', d[0]))
    
def extract_title(x):
    """
    Extracts title from title column of poster dataset.
    """
    t = re.sub('\(\d{4}\)', '', x)
    return t.strip().lower()

poster = pd.read_csv(poster_path)
summ = pd.read_csv(summ_path)

# Clean title columns
poster["Date"] = poster["Title"].apply(extract_date)
poster["title"] = poster["Title"].apply(extract_title)

summ["Title"] = summ["Title"].apply(lambda x: x.strip().lower())

# Merge two datasets
data = pd.merge(poster, summ, left_on = ["title", "Date"], right_on = ["Title", "Release Year"], how = "inner")

# Drop repetitive rows
data = data.drop(["Title_x", "Genre_y", "Title_y", "Date"], axis = 1)

# Rename columns
data = data.rename(columns={"Genre_x": "Genre", "title": "Title"})
print("Data Shape: {}".format(data.shape))
data.head(1)
```

### Extracting Colors
Now we need to make our dataset practical for our research question. To do that, we have to load the posters and extract salient colors from the posters. Define functions for extracting color from image. 

```python
def get_colors(img_url, tolerance, num_colors):
    """
    Returns list of colors featured in an image.
    
    :param img_url: url to image
    :param tolerance: rounds RGB value to nearest tolerance. A low tolerance will result in a color palette with variations of the same color
    :param num_colors: number of colors to output
    
    :return: list of RGB tuples
    """
    try:
        img = Image.open(requests.get(img_url, stream=True).raw)
    except:
        return None
    img_color = defaultdict(int)
    round_base = lambda x: min([tolerance * round(x/tolerance), 255])
    for px in img.getdata():
        px_rounded = (round_base(px[0]), round_base(px[1]), round_base(px[2]))
        img_color[px_rounded] += 1
    palette = sorted(img_color, key=lambda x:img_color[x], reverse=True)[:num_colors]
    return palette
```

```python
def show_palette(palette):
    """
    Prints color palette.
    
    :params palette: list of RGB tuples
    
    :return: none
    """
    if palette == None:
        return
    palette = np.array(palette)[np.newaxis, :]
    plt.imshow(palette);
    plt.axis('off');
    plt.show();
```

We first visualized the top colors for each genre but realized that our values are quite dull. Therefore, we decided to do further cleaning. The first method we tried was eliminating neutral colors. Greys, blacks, and whites should be taken out of priority and chosen only if no other choice is available.

An important note was that is was clear that black and white were the outliers of our data. These outliers made too much noise so we couldn't see the actual variance of colors and because of this, we implemented the "No_Neutral_Color" dataframe.

```python
def detect_neutral(color):
    neutral = False
    if (color[0] == color[1] and color[1] == color[2] or # detects same RGB values
        color[0] >= 210 and color[1] >= 210 and color[2] >= 210 or # detects whites
            color[0] <= 60 and color[1] <= 60 and color[2] <= 60): # detects blacks
        neutral = True
    return neutral
```

```python
def ignore_neutral_colors(palette, color):
    i = 0
    while detect_neutral(color) and i < 9:
        i += 1
        color = palette[i]
    if detect_neutral(color) and i == 9:
        color = palette[0]
    return color
```

```python
df["No_Neutral_Color"] = df.apply(lambda df: ignore_neutral_colors(df["Palette"], df["Color"]), axis=1)
```

### K-means Clustering Regression Hybrid

In this portion of our project, we are utilizing a k-means clustering model that takes in sentiment and genre data and plots clusters in which we hoped we would be able to predict a color palette for a movie poster that is the same as the poster's real color palette.

```python
genres['plot_neg'] = df['plot_neg']
genres['plot_neu'] = df['plot_neu']
genres['plot_pos'] = df['plot_pos']
```

Since we will be using our 'genres' dataframe for this K-means model, we decided to include data from our sentiment analysis in order to generate a more accurate representation of our data. We thought our sentiment analysis would be able to help in creating a color palette by bringing in an aspect of how colors evoke certain emotions.

```python
y_df = pd.DataFrame()
y_df["R"] = df["No_Neutral_Color"].apply(lambda x: x[0])
y_df["G"] = df["No_Neutral_Color"].apply(lambda x: x[1])
y_df["B"] = df["No_Neutral_Color"].apply(lambda x: x[2])

cols = ['Action', 'Adventure', 'Animation', 'Biography',
       'Comedy', 'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy',
       'Film-Noir', 'History', 'Horror', 'Music', 'Musical', 'Mystery', 'News',
       'Romance', 'Sci-Fi', 'Short', 'Sport', 'Talk-Show', 'Thriller', 'War',
       'Western', 'plot_neg', 'plot_neu', 'plot_pos']
X_df = df[cols]

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.33, random_state=42)
```

In the above cell we made sure to set our x and y dataframes that we will use in a train test split. The train test split is used to fit our K-means model with the train dataset and then create predictions from this with our test dataset. 

```python
num_clusters = 10

X_train = X_train.loc[~X_train.isnull().any(axis=1)]
val = X_train.values
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(val)
y_kmeans = kmeans.predict(val)
X = val
pca = PCA(n_components=4)
decomVal = pca.fit_transform(X)
```

In this part of the K-means model we are fitting our K-means model with the train dataset. We set our clusters to 10 expecting to find 10 clusters in our data visualization. Then we had to perform a principal component analysis (PCA) in order to reduce the dimensions of our data because we weren't able to plot as many dimensions as our data had. We hoped that the PCA would reduce noise and allow our model to become more efficient and preserve signal.

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(decomVal[:, 0], decomVal[:, 1], decomVal[:, 2], c=y_kmeans, s=50, cmap='Pastel2')

centers = kmeans.cluster_centers_
ax.scatter3D(centers[:, 0], centers[:, 1], centers[:, 2], c='black', s=200, alpha=0.5)
```

Here we are generating a 3D scatterplot of our K-means model. Several clusters can be identified here through the different pastel colors assigned to each one. We also plotted centroids for these clusters that gave us an idea of where many of these points lie from each cluster as each centroid is the center (arithmetic mean) of a cluster.

Now we will train our k-means regression hybrid model. To do this, we will first determine which cluster each point is in. Then, we will fit a regression line in each cluster. Finally, we will predict RGB values according to cluster.

```python
models = []

pred = kmeans.predict(X_train)
X_train["clusters"] = pred

for i in range(num_clusters):
    linear = linear_model.LinearRegression()
    cluster_y = y_train.loc[X_train["clusters"] == i]
    linear.fit(X_train.loc[X_train["clusters"] == i].drop("clusters", axis=1), cluster_y)
    models.append(linear)

X_test["clusters"] = kmeans.predict(X_test)
predictions = pd.DataFrame()
for i in range(num_clusters):
    clust_df = X_test.loc[X_test["clusters"] == i].drop("clusters", axis=1).copy()
    clust_y = y_test.loc[X_test["clusters"] == i]
    clust_df[["R_Pred", "G_Pred", "B_Pred"]]  = models[i].predict(clust_df)
    clust_df = clust_df.merge(clust_y, left_index=True, right_index=True)
    predictions = pd.concat([predictions, clust_df])
```

## Palette Prediction
### Generative Adversarial Networks with Fully-Connected Generator
While training our color prediction models, we also trained models to predict palettes. Because we knew from the beginning that this task would require a complex model, we trained a neural network. Partly based on [this](https://towardsdatascience.com/build-a-super-simple-gan-in-pytorch-54ba349920e4) model, heavily inspired by [this](https://openaccess.thecvf.com/content_ECCV_2018/papers/Hyojin_Bahng_Coloring_with_Words_ECCV_2018_paper.pdf) paper.

In order to make believable palettes, we found in our research that the best neural network to use would be GAN. We tried using several different features, but found that using word embeddings had the best performance. This is largely due to many of our useful features being categorical, making our input data sparse. In order to use those features, we would have to add a considerable amount of random noise to 0s to avoid a gradient vanishing problem, which made our inputs essentially random. So, we used fasttext pretrained word embeddings.

```python
import torch
from torch.utils.data import Dataset, DataLoader, random_split
import torch.nn as nn
from torch import optim

import fasttext.util
import colorsys
# fasttext.util.download_model('en', if_exists='ignore')

ft = fasttext.load_model('data/raw/cc.en.300.bin')
# Reduce dimension to 100
fasttext.util.reduce_model(ft, 100)
```

First, we set up our input and output values by embedding the first 200 words in the plot column normlizing the 0-255 RGB values to be between -1 and 1.

```python
# Embedding inputs
def embedding(ls):
    emb = [ft.get_word_vector(word).tolist() for word in ls][:200]
    if len(emb) < 200:
        num_zeros = 200 - len(emb)
        emb = emb + [np.random.normal(0, 0.5, 100).tolist() for word in range(num_zeros)]
    return emb

plots_emb = df["Plot"].apply(embedding)
inp = np.array(plots_emb.tolist())

# Normalizing outputs
def normalize(x):
    return x/256-1

out = [[[normalize(k) for k in list(c)] for c in x] for x in df["Palette"]]
```

Then, we put our data into a dataloader and split our data into train/test sets.

```python
class PaletteDataset(Dataset):
    def __init__(self, inp, out, transform=None):
        super(PaletteDataset, self).__init__()
        self.transform = transform
        self.X = inp
        self.Y = out
        
    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return [np.array(self.X[idx]), np.array(self.Y[idx])]
```

```python
batch_size = 32
dataset = PaletteDataset(inp, out)
dataloader = DataLoader(dataset, batch_size=batch_size, drop_last=True, shuffle=True)

test_size = int(len(dataset)*0.3)

train_dataset, val_dataset = random_split(dataset, (len(dataset)-test_size, test_size))

train_loader = DataLoader(train_dataset, batch_size=batch_size, drop_last=True, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, drop_last=True, shuffle=True)
```

Subsequently, we build our model.

```python
# Define device to do computations
device = torch.device("cuda:0" if (torch.cuda.is_available() and ngpu > 0) else "cpu")
```

We make a two linear layer generator.

```python
class Generator(nn.Module):
    def __init__(self, inp_dim: int, out_dim: int):
        super(Generator, self).__init__()
        self.fc1 = nn.Linear(inp_dim, out_dim)
        self.fc2 = nn.Linear(out_dim, out_dim)
        self.tanh = nn.Tanh()

    def forward(self, x):
        x = self.fc1(x)
        x = self.tanh(self.fc2(x))
        return x
```

And a simple discriminator.

```python
class Discriminator(nn.Module):
    def __init__(self, inp_dim: int):
        super(Discriminator, self).__init__()
        self.fc1 = nn.Linear(inp_dim, 1);
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        return self.sigmoid(self.fc1(x))
```

Then, we define our training function that sets up the architecture of a GAN.

```python
def train(generator, 
          discriminator, 
          device, train_loader, 
          generator_criteron, discriminator_criterion,
          generator_optimizer, discriminator_optimizer, epoch, log_interval=10000):
    iterator = tqdm(train_loader, total=int(len(train_loader)))

    
    for batch_idx, (data, target) in enumerate(iterator):
        
        generator_optimizer.zero_grad()
        
        data = torch.reshape(data, (batch_size, -1)).float().to(device)
        target = torch.reshape(target, (batch_size, -1)).float()
        
        generated_data = generator(data)
        true_labels = torch.ones(batch_size).float()
        true_data = target.float()

        fake_labels = torch.zeros(batch_size).float()
        
        # Measures how well discriminator was tricked
        generator_discriminator_out = discriminator(generated_data).flatten()
        generator_loss_gan = discriminator_criterion(generator_discriminator_out, true_labels)
        

        generator_loss_smoothL1 = generator_criteron(generated_data, target)
        
        # Generator loss consists of both how badly discriminator has done 
        # and how close our predicted values are to the actual labels
        generator_loss = generator_loss_gan + generator_loss_smoothL1
        
        generator_loss.backward()
        generator_optimizer.step()
        
        # Train discriminator on actual values
        discriminator_optimizer.zero_grad()
        
        true_discriminator_out = discriminator(true_data).flatten()
        true_discriminator_loss = discriminator_criterion(true_discriminator_out, true_labels)
        
        
        generator_discriminator_out = discriminator(generated_data.detach()).flatten()
        generator_discriminator_loss = discriminator_criterion(generator_discriminator_out, fake_labels)
        
        discriminator_loss = (true_discriminator_loss + generator_discriminator_loss) / 2
        discriminator_loss.backward()
        discriminator_optimizer.step()    
        
        iterator.set_postfix(discriminator_loss = discriminator_loss.item(), generator_loss = generator_loss.item())
```

Now, we train! We will be using Adagrad for our generator optimizer because our model tended to fluctuate losses a lot. We hoped that it would be able to help adjust our learning rate, but the results were only decent. We used SmoothL1Loss to judge our generator as proposed by the research paper we followed.

```python
from tqdm.notebook import tqdm

input_dim = 20000
output_dim = 30
num_epoch = 100

generator = Generator(input_dim, output_dim)
generator = generator.to(device)

discriminator = Discriminator(output_dim)
discriminator = discriminator.to(device)

generator_optimizer = torch.optim.Adagrad(generator.parameters(), lr=0.00001)
discriminator_optimizer = torch.optim.SGD(discriminator.parameters(), lr=0.00001)

generator_loss = nn.SmoothL1Loss()
discriminator_loss = nn.BCELoss()


for epoch in range(1, (num_epoch + 1)):
    train(generator, discriminator, device, train_loader, generator_loss, discriminator_loss, generator_optimizer, discriminator_optimizer, epoch)
```

# Conclusion & Discussion

The research we gathered was focused on the correlation between human emotions and color palettes within films from all around the world. At first, our team hypothesized that we will be able to identify relationships between colors and characteristics of a movie. More specifically, we believed that warmer color palettes on movie posters would evoke positive emotions while cooler color palettes would evoke negative emotions. 

Since the accuracy of color palettes/representative colors is hard to quantify, we used the help of line plots, barcharts, 3D scatterplots, and other visusalization tools to visually display the correlations between colors and our features. To analyze our data, we cleaned our data and ran numerous tests to understand certain specifics of our data including metrics such as 'Top 20 Colors of Bollywood Comedies' and 'Top 20 Colors of Horror Films from 1990'. Based on our EDA, we found optimistic results corroborating our initial hypothesis and gained insight on how to go about training our model. 

One piece of important information we learned is that, based on way we extracted the colors from the poster, neutral colors (specifically blacks, whites, and greys) were overrepresented as  the most salient color in each movie. As a result of this overrepresentation, plotting the most representative color of a movie based on our features, such as genre, yielded mostly neutral colors. It was only when we looked at specific subsets of the data that we could find meaningful results. This understanding would later become our inspiration to do a K-Means regression hybrid model.

Another important piece of information gained from our exploratory data analysis was that utilizing all of the extracted colors is fundamental. While plotting only single representative colors based on genre yielded unsatisfactory results, plotting all of the extracted colors based on genre, we were clearly able to see the differences between the genres through colors. For example, horror movies had substantially more darker colors than romance movies. With this new information, we began to make models while focusing on genre.

We trained three models first: linear regression, k-means regression, hybrid, and polynomial. Unfortunately, these models yielded very neutral predictions, likely because minimizing the mean squared error equated to choosing values close to the mean for "R", "G", and "B" values. In addition, the predicted values were not in reference to each other. Realizing that any form of model that required fitting a line through our data would not work, we decided to train a simple k-nearest neighbor model. Visually, the results for the k-nearest neighbor model were the best, while the linear regression model had the best quantitive results. Simultaneously, we trained a similarly disappointing GAN model that predicted color palettes. Although the color palettes were aesthetically pleasing, our model ran into a mode collapse, and started creating only a small range of outputs.

Each model we trained was successful in one facet of our overall goal to predict color palettes/representative colors for a movie (i.e. k-nearest neighbor for believable results, linear regression for lowest quantitative loss, and palettes for aesthetically pleasing results), but we weren't able to find a model that was able to comprehensively incorporate everything into a color palette. We also weren't able to quantitatively determine that warmer and cooler colors were linked to certain emotions. Although our findings weren't what we expected, we were able to visually display relationships between color, genre, origin through our EDA. We learned that linear models would do poorly on color data as it attempts to minimize MSE so it picks average values. Non-linear models do better, but simple models are not good enough. So in the future, we will train an RNN-based model as our generator in the future as it is more fitting for the problem we have. Despite our failures, we were still able to succeed in making relatively good baseline models to answer our difficult question.
