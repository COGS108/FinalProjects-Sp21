```python
#Next, we will use dropna to drop all the null values.
df_doge_yahoo = df_doge_yahoo.dropna()
df_doge_gecko = df_doge_gecko.dropna()
df_bitcoin_yahoo = df_bitcoin_yahoo.dropna()
df_bitcoin_gecko = df_bitcoin_gecko.dropna()
df_doge_btc_gecko = df_doge_btc_gecko.dropna()
```

```python
#We also find out that dataset from coingecko has too many observations that we don't need. We then decide that we will cut it then reorder the dataset.
df_doge_gecko = df_doge_gecko[871:];
df_bitcoin_gecko = df_bitcoin_gecko[871:];
df_doge_btc_gecko = df_doge_btc_gecko[803:];

#Now we will reset the index for each dataset.
df_doge_gecko = df_doge_gecko.reset_index(drop=True)
df_doge_yahoo = df_doge_yahoo.reset_index(drop=True)
df_bitcoin_gecko = df_bitcoin_gecko.reset_index(drop=True)
df_bitcoin_yahoo = df_bitcoin_yahoo.reset_index(drop=True);
df_doge_btc_gecko = df_doge_btc_gecko.reset_index(drop=True);
```

```python
#Since we only need market_cap column for the coingecko datasets, we will combine the datasets
df_doge_marketCap = df_doge_gecko[['market_cap']]
df_bitcoin_marketCap = df_bitcoin_gecko[['market_cap']]

#Now, we will combine the dataset to one
dogecoin = pd.concat([df_doge_yahoo, df_doge_marketCap], axis=1, join='inner')
bitcoin = pd.concat([df_bitcoin_yahoo, df_bitcoin_marketCap], axis=1, join='inner')
```

```python
#Function to standardize date in doge-btc pricing dataset
def standardize_date(str_in):
    
    try:
        str_in = str_in.lower()
        str_in = str_in.strip()

        if '00:00:00 utc' in str_in:
            str_in = str_in.replace('00:00:00 utc','')
            output = str_in
        else:
            output = str_in
    except:
        output = str_in
        
    return output
```

```python
df_doge_btc_gecko = df_doge_btc_gecko.rename(columns={'snapped_at':'Date', 'price': 'BTC-Valuation','total_volume':'Volume'})
df_doge_btc_gecko['Date'] = df_doge_btc_gecko['Date'].apply(standardize_date)

#Renaming for consistency with other datasets, as well as standardizing dates
df_doge_btc_gecko.head()
```
