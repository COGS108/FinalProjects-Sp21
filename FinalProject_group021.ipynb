## Background & Prior Work

Life expectancy estimates the average number of years a person can expect to live. While factors such as gender, genetics, hygiene, diet, and exercise are known to affect life expectancy, many data scientists discovered more factors that influence life expectancy.

Dr. Lei and Na investigated the relationship between life expectancy and economic growth for 65 countries and found significantly positive relationships in most countries. More specifically, this relationship is stronger in groups with higher levels of aging [1]. 

On the other hand, Wolfgang Lutz found that education has a more significant impact on life expectancy than the economy, because education improves people's cognitive abilities which allow them to make better choices throughout the rest of their life [2]. By analyzing changes of life expectancy in Italy, Denmark, and the U.S. between 1990 and 2010, another research conducted by Marc Luy also concluded that improvements in longevity can be attributed to increasing education levels. The contribution of changes in the population educational structure to improvement in life expectancy is ~15% for men in the USA and ~40% for women in Denmark [3]. 

Moreover, according to data collected on government expenditure on healthcare and life expectancy between 1970 and 2015 in a number of developed countries, the population lives longer as health expenditure increases [4]. Similar results were found in eight East African countries [5]. However, this trend is much weaker in the U.S. compared to in other countries. Some reasons for this underachievement may be the large inequality in health spending and high levels of violence in the U.S. [4]. 

Our project looks into the relationship between life expectancy and government expenditure on health care, economic level, income composition of resources, and schooling. By analyzing data from 193 countries, we hope to determine their association with life expectancy on a global scale.

```python
gdpdf = pd.read_csv("gdp.csv")
gdp = gdpdf.drop(["Indicator Name", "Indicator Code"] + [str(i) for i in list(range(1960, 2000))], axis = 1) # Unused Columns Dropped
gdp = gdp.melt(id_vars=['Country Name'], value_vars = [str(i) for i in list(range(2000, 2016))]) # Reorder the structure

gdp = gdp.rename(columns = {"Country Name": "Country", "variable": "Year", "value": "cleaned GDP"})

# Rename countries to match dataset 
gdp = gdp.replace("Yemen, Rep.", "Yemen").replace("Vietnam", "Viet Nam") 
gdp = gdp.replace({"Bahamas, The": "Bahamas", "Bolivia": "Bolivia (Plurinational State of)", "Venezuela, RB": "Venezuela (Bolivarian Republic of)", 
                  "United States": "United States of America", "Cote d'Ivoire": "Côte d'Ivoire", "Tanzania": "United Republic of Tanzania", 
                  "Congo, Dem. Rep.": "Democratic Republic of the Congo", "Czech Republic": "Czechia", "United Kingdom": "United Kingdom of Great Britain and Northern Ireland", 
                  "North Macedonia": "The former Yugoslav republic of Macedonia", "Congo, Rep.": "Congo", "Gambia, The": "Gambia", 
                  "Lao PDR": "Lao People's Democratic Republic", "St. Lucia": "Saint Lucia", "Micronesia, Fed. Sts.": "Micronesia (Federated States of)", 
                  "Kyrgyz Republic": "Kyrgyzstan", "Iran, Islamic Rep.": "Iran (Islamic Republic of)", "Egypt, Arab Rep.": "Egypt", 
                  "Slovak Republic": "Slovakia", "St. Vincent and the Grenadines": "Saint Vincent and the Grenadines", "Moldova": "Republic of Moldova", 
                  "Korea, Rep.": "Republic of Korea", "Eswatini": "Swaziland"}) 
gdp["Year"] = gdp["Year"].astype("int") # String to int for merge
df_clean = df_clean.merge(gdp, how = "left", on = ["Country", "Year"]) 

# Northern Korea and Somalia has very limited information, we drop those countries from dataset
df_clean = df_clean[~df_clean["Country"].isin(["Democratic People's Republic of Korea", "Somalia"])]
df_clean.isnull().sum() / len(df_clean)
```

The cleaned GDP column only has <1% values missing! We can drop the original GDP columns.

We have 2 approaches to impute the missing values within other variables. 
1. impute using mean or median of the entire dataset. 
2. collect those information from other dataset. 

While approach 2 might be much more accurate than approach 1, it requires a lot of additional manual searching and replacing, which is too expensive for the scope of this project. Also there is very few values missing in our dataset after we cleaned GDP, on average only around 5% data are missing, which are not quite significant. We will use the first approach. We use histograms to decide which value (mean or median) to use for imputation. 

```python
pd.plotting.scatter_matrix(df_clean[["Alcohol", "Total expenditure", "GDP", "Income composition of resources", "Schooling"]], figsize = (27, 27))
print()
```

It seems Alcohol, Total expenditure and GDP are skewed right a lot. We believe it's better to impute using median. 

While Schooling is generally in gaussian distribution. We will fill the missing schooling data with mean values. 

"Income composition of resources" doesn't seemed to be normally distributed, we fill it with median values. 

```python
df_clean['Alcohol'].fillna((df_clean['Alcohol'].median()), inplace = True)
df_clean['Total expenditure'].fillna((df_clean['Total expenditure'].median()), inplace = True)
df_clean['GDP'].fillna((df_clean['GDP'].median()), inplace = True)

df_clean['Income composition of resources'].fillna(value = df['Income composition of resources'].median(), inplace = True)

df_clean['Schooling'].fillna(value=df['Schooling'].mean(), inplace = True)
```

## Can we predict life expectancy based on the income/composition of resources?

We have observed that there is a rather strong correlation between life expectancy and income/composition of resources in our visualiztions. Thus, we can generate a linear regression model using our data and assess this model to see if this factor can be used to predict life expantancy. But as we take a closer look at our dataset, it seems like there are quite a few countries with income/composition of resources of 0 in our dataset. This occurs not only in countries with lower life expectancy, but also countries with life expectancy around 80. So we did some background research, and found out that income/composition of resources was never 0 in any country over the past 30 years, so the 0.0s are likely indicating absence of data. So we decided to fill in those missing values with data collected from http://hdr.undp.org/en/indicators/103606#, a website of the United Nations Development Programme

```python
df2 = pd.read_csv('Income index.csv',encoding='latin-1')
df2.head(3)
```

Merge two datasets. 

```python
df_clean = df_clean.rename(columns = {'Income composition of resources':'Income_composition_of_resources', 'Life expectancy':'Life_expectancy',  
                                      'Total expenditure' : 'Total_expenditure', 'percentage expenditure' : 'percentage_expenditure'}) # We replaced space in column names with '_' to make data processing easier.
df2 = df2.drop('HDI Rank', 1) # We will drop the HDI Rank column since it's not relevant for our task. 

# Convert the year columns in df2 to a single single column.

df2 = df2.melt(id_vars='Country', var_name='Year', value_name='Income composition of resources').sort_values(by=['Country','Year']).reset_index(drop=True)
df_clean['Income_composition_of_resources'] = df_clean['Income_composition_of_resources'].replace(0, np.nan) # 0.0 values in the Income/composition of resources column is replaced by NaN 

# Prepare df2 merging
df2['Year'] = df2['Year'].astype(int)
df2['Country'] = df2['Country'].str.strip() 

# Merge 
merge = pd.merge(df_clean, df2, on=['Country','Year'], how='left')
merge['Income_composition_of_resources'] = merge['Income_composition_of_resources'].fillna(merge['Income composition of resources'])
merge = merge.drop('Income composition of resources', axis=1) 

merge = merge.dropna() # We will drop the rows that cannot obtain any value from either of our datasets
merge.head(3)
```

#### Modeling 
Data is splited into training and testing with the ratio of 8:2

```python
x_train, x_test, y_train, y_test = train_test_split(merge['Income_composition_of_resources'], merge['Life_expectancy'], test_size=0.2, shuffle=False)
df_train = pd.concat([x_train , y_train], axis = 1) 
df_test = pd.concat([x_test , y_test], axis = 1) 

# Generate a linear regression model using our data
outcome, predictor = patsy.dmatrices('Life_expectancy ~ Income_composition_of_resources	', df_train)
mod = sm.OLS(outcome, predictor)
res = mod.fit() 

print(res.summary())
```

To see the our model's performance on test data, we calculated its RMSE

```python
predict = res.predict(sm.add_constant(x_test))
mean_squared_error(predict, y_test)
```

The R-squared value is at 0.751, which indicates a good fit, and the p-value is 0.000, which is smaller than the significance level we use (0.05), which means we can reject the null hypothesis. This also proves that our assumption is correct: there is a rather strong correlation bewteen Income/composition of resources and life expectancy, and it's not caused by random chance. We can then extract our model: 

$$ Life  Expectancy = 36.2381 + 50.4472  * Income Composition of Resources $$

Which indicates people without any Income Composition of Resources would have a life expectancy of 36.2381, and every one percent increase in Income Composition of Resources leads to 0.5 years of increase of life expectancy. Note that this model only indicates a correlation bewteen the variables and does not imply causation.

We fit a line to the scatterplot.

```python
plt.figure(figsize=(12,12))
sns.scatterplot(y = merge['Life_expectancy'],x = merge['Income_composition_of_resources'], data=merge)
plt.title("Life Expectancy with respect to Income composition of resources",fontsize = 20)

xs = np.array([merge['Income_composition_of_resources'].min(), merge['Income_composition_of_resources'].max()])
ys = 51.0412 * xs + 35.7384 
plt.plot(xs, ys, '-', linewidth=4, color = "red", label='Model')
plt.legend();
```

By looking at our model on the scatter plot and the standard error, we came to a conclusion that the income/composition of resources is in fact an index that can be used to predict life expectancy of a country.

## Relationship between Life expectancy and economics

Since GDP is a way to measure the economics level of a country, We decided to use the linear regression model to illustrate the relationship between GDP and life expectancy. As the GDP plot indicates, the GDP distribution is highly skewed. Thus, we conducted log transformation on the GDP to alleviate the skewness. By simpliliy observing the heapmap and the scatter plot between log transformed GDP and life expectancy, we found there exist a moderate positive relationship. There is no 0 value in the GDP column, so we are not going to use a secondary dataset. 

#### Modeling 
Data is splited into training and testing with the ratio of 8:2

```python
x_train, x_test, y_train, y_test = train_test_split(df_clean['GDP'], df_clean['Life_expectancy'], test_size=0.2, shuffle=False)
df_train = pd.concat([x_train , y_train], axis = 1) 
df_test = pd.concat([x_test , y_test], axis = 1) 

# Model 
outcome, predictor = patsy.dmatrices("Life_expectancy ~ np.log(GDP)", df_train)
mod = sm.OLS(outcome, predictor)
res = mod.fit()

print(res.summary())
```

To see the our model's performance on test data, we calculated its RMSE

```python
predict = res.predict(sm.add_constant(np.log(x_test)))
mean_squared_error(predict, y_test)
```

According to the result, the R-squared value is at 0.645 and the p-value is 0.000. It indicates a good fit and the result is statistically significant (0 < 0.05). We can reject the null hypothesis. There exist positve relationship between economics and Life Expectancy. We can then extract our model:

$$ Life  Expectancy = 46.1890  + 3.1064 * log(GDP) $$

Which indicates people in country of 0 GDP would have a life expectancy of 46.189, and every one unit increase in log(GDP) leads to 31 years of increase of life expectancy. Note that this model only indicates a correlation bewteen the variables and does not imply causation. 

```python
plt.figure(figsize=(12,12))
sns.scatterplot(y = df_clean['Life_expectancy'],x = np.log(df_clean['GDP']), data=merge)
plt.title("Life Expectancy with respect to economics",fontsize = 20)

xs = np.array([np.log(df_clean['GDP']).min(), np.log(df_clean['GDP']).max()])
ys = 3.0714 * xs + 46.2846
plt.plot(xs, ys, '-', linewidth=4, color = "red", label='Model')
plt.legend();
```

## Predict life expectancy based on the quality of medical treatment & health care

It is hard to establish a accurate stanard to measure the quality of medical treatment & health care for a country in a single year. After conducting some research, we believe government and individual's spending on the health care/technology are significant factors. Thus, we decided to specify quality of medical treatment & health care as the combination of government and the individual expenditure on health care. However, our original dataset does not include data about government expenditure per year. To access that information, we found another dataset from https://ourworldindata.org/government-spending and merged it into our original dataset. 

```python
df3 = pd.read_csv('total-gov-expenditure-gdp-wdi.csv',encoding='latin-1')
df3 = df3.drop('Code', axis = 1) # Remove unused column and renamed columns to simplify data processing
df3['Entity'] = df3['Entity'].str.strip() # Clean Entity

df3 = df3.rename(columns = {'Entity' : 'Country', "Expense (% of GDP)" : "Government_expense_percentage"})

merge2 = pd.merge(df_clean, df3, on=['Country','Year'], how='left') # Merge dataset
merge2 = merge2.dropna()
merge2.head(3)
```

**Feature Engineering** 

We added two columns: Government_expense_health_total and Individual_expense_health_total <br>
- Government_expense_health_total: represents the government total spending on health care per capita. It is calculated base on the formula: government expenditure (% of GDP) * GDP * government expenditure percentage on health care <br>
- Individual_expense_health_total: represents the individual's total spending on health care. It is calculated base on the formula: individual expenditure percentage on health care* GDP

```python
merge2['Government_expense_health_total'] = merge2['Government_expense_percentage']  * merge2['GDP'] * merge2['Total_expenditure'] / 100 / 100
merge2['Individual_expense_health_total'] = merge2['GDP'] * merge2['percentage_expenditure'] / 100
```

Before we conduct the regression, we need to take a look at the distribution of our data. As shown in the graph below, government expenditure respect to life expectancy is highly skewed right, which suggests a log transformation. Similarly, the distirbution of individual expenditure of health care is also highly skewed right. However, in this case, there exist data with value 0. Since log(0) is undefined, we replace all 0 with 0.000001 which is smaller than any other value in the column. 

```python
sns.set(rc = {'figure.figsize':(7, 7)})
sns.histplot(x = "Government_expense_health_total", data = merge2)
plt.title("Distribution of Government_expense_health_total",fontsize = 20)
```

```python
merge2["Individual_expense_health_total"] = merge2["Individual_expense_health_total"].replace(0, 0.000001)
```

#### Modeling 
Data is splited into training and testing with the ratio of 8:2

```python
x1_train, x1_test, x2_train, x2_test, y_train, y_test = train_test_split(merge2['Government_expense_health_total'], merge2['Individual_expense_health_total'], merge2['Life_expectancy'], test_size=0.2, shuffle=False)
df_train = pd.concat([x1_train ,x2_train,  y_train], axis = 1) 
df_test = pd.concat([x1_test ,  x2_test , y_test], axis = 1) 

# Model 
outcome, predictor = patsy.dmatrices("Life_expectancy ~  Individual_expense_health_total + np.log(Government_expense_health_total) ", df_train)
mod = sm.OLS(outcome, predictor)
res = mod.fit()

print(res.summary())
```

To see the our model's performance on test data, we calculated its RMSE 

```python
predict = res.predict(sm.add_constant(pd.concat([x2_test, np.log(x1_test)], axis = 1) ))
mean_squared_error(predict, y_test)
```

According to the result, the R-squared value is at 0.578 and the p-value is 0.017 and 0.000, both smaller than the significance level we use (0.05). We can reject the null hypothesis. There exist positve relationship between economics and government expenditure on health care. We can then extract our model: 

    Life  Expectancy = 57.7796  + 0.0001 * Individual_expense_health_total + 3.3042 * log(Government expenditure on health care)

People with 0 total individual expense on health and 0 government expenditure on health has life expectancy of 57.7796. Every one dollar increase in total individual expense on health increases life expectancy by 0.0001, every one unit increase in log(government expenditure on health) increases life expectancy by 3.3042. 
Note that this model only indicates a correlation bewteen the variables and does not imply causation

### Predict using all dataset variables 

Above analysis has shown those variables can be used for prediction. We use all variables in our dataset to predict Life Expectancy. 

```python
from sklearn.linear_model import LinearRegression

# Create and transform dataset 

allva = df_clean.copy()

allva = allva.merge(merge[["Year", "Country", "Income_composition_of_resources"]], on = ["Year", "Country"], how = "left").dropna()
allva = allva.drop(["Country", "Income_composition_of_resources_x"], axis = 1) # We don't want country to be predictor 

allva["GDP"] = np.log(allva["GDP"]) # Log transform GDP 
allva["percentage_expenditure"] = np.log(allva["percentage_expenditure"]).replace(-np.inf, -5.7) # Log transform percentage expenditure and replace -infinity lowest value -6 

allva = allva.replace({"Developing": 0, "Developed": 1}) # Encode status 

# Train Test Split 
X_train, X_test, y_train, y_test = train_test_split(allva.drop("Life_expectancy", axis = 1), allva[["Life_expectancy"]], test_size = 0.2, random_state = 42)

# Model 
lr = LinearRegression().fit(X_train, y_train)

# RMSE 
print("RMSE on Training: " + str(mean_squared_error(lr.predict(X_train), y_train)))
predict = lr.predict(X_test)
print("RMSE on Testing: " + str(mean_squared_error(predict, y_test)))
```

Using all variables, the resulting model has RMSE smaller than any model we built above! This makes sense because any additional information (if it's not noise) usually helps. 

This simple Linear Regression model is our baseline model, we can build something more complex, for example, using Random Forest Regressor and Cross Validation. 

```python
from sklearn.model_selection import GridSearchCV
import warnings
from sklearn.ensemble import RandomForestRegressor
warnings.filterwarnings('ignore')

# Cross Validation for best parameter
parameters = {"n_estimators": [50, 500, 1000], "max_features": [0.7, None], "max_depth": [12, 30, None]}
reg = RandomForestRegressor(n_jobs = -1)
clf = GridSearchCV(reg, parameters).fit(X_train, y_train)

# Modeling 
reg = RandomForestRegressor(n_estimators = clf.best_params_["n_estimators"], max_features = clf.best_params_["max_features"], 
                            max_depth = clf.best_params_["max_depth"], n_jobs = -1)
reg = reg.fit(X_train, y_train)

# RMSE 

print("RMSE on Training: " + str(mean_squared_error(reg.predict(X_train), y_train)))
predict = reg.predict(X_test)
print("RMSE on Testing: " + str(mean_squared_error(predict, y_test)))
```

Using Cross Validation and Random Forest Regressior, we built a complex model which reduces 95% training RMSE and over 70% testing RMSE! The testing RMSE is significantly higher than the training RMSE though, which indicates overfitting and suggests furtuer tunning is required.
